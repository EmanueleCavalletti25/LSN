### Overview 

In this notebook our task will be to perform machine learning regression on noisy data with a Neural Network (NN).

We will explore how the ability to fit depends on the structure of the NN. The goal is also to build intuition about why prediction is difficult.

### The Prediction Problem

Consider a probabilistic process that gives rise to labeled data $(x,y)$. The data is generated by drawing samples from the equation

$$
    y_i= f(x_i) + \eta_i,
$$

where $f(x_i)$ is some fixed, but (possibly unknown) function, and $\eta_i$ is a Gaussian, uncorrelate noise variable such that

$$
\langle \eta_i \rangle=0 \\
\\
\langle \eta_i \eta_j \rangle = \delta_{ij} \sigma
$$

We will refer to the $f(x_i)$ as the **true features** used to generate the data. 

To make predictions, we will consider a NN that depends on its parameters, weights and biases. The functions that the NN can model respresent the **model class** that we are using to try to model the data and make predictions.

To learn the parameters of the NN, we will train our models on a **training data set** and then test the effectiveness of the NN on a *different* dataset, the **validation data set**. The reason we must divide our data into a training and test dataset is that the point of machine learning is to make accurate predictions about new data we have not seen.

To measure our ability to predict, we will learn our parameters by fitting our training dataset and then making predictions on our test data set. One common measure of predictive  performance of our algorithm is to compare the predictions,$\{y_j^\mathrm{pred}\}$, to the true values $\{y_j\}$. A commonly employed measure for this is the sum of the mean square-error (MSE) on the test set:
$$
MSE= \frac{1}{N_\mathrm{test}}\sum_{j=1}^{N_\mathrm{test}} (y_j^\mathrm{pred}-y_j)^2
$$

We will try to get a qualitative picture by examining plots on validation and training data.

### Linear fit

We start by considering the very simple case:
$$
f(x)=2x+1
$$

Let's start defining the parameters of an ideal linear function which we are going to predict through a neural network regression

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

Now let's generate a set of input data which will slightly deviate from our ideal behaviour using a random noise (that actually is set to zero):

```python
import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 500)
x_valid = np.random.uniform(-1, 1, 50)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
# plot validation and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target, label='target')
plt.scatter(x_valid, y_valid, color='r', label='validation data')
plt.legend()
plt.grid(True)
plt.show()
```

If you remember how a single node of a neural network works, you can easily spot that **just a single neuron can make the job**. So let's start using a simple Sequential model with just one layer on one neuron only!

questo perchè ho il peso che moltiplica, l'ascissa e poi aggiunge il termine noto

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

# Definisci il modello sequenziale
model = Sequential([
    Input(shape=(1,)),  # Definisci la forma dell'input come livello separato
    Dense(1)            # Aggiungi il livello denso
])

# Compila il modello scegliendo l'ottimizzatore, la funzione di perdita e le metriche
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# Stampa un riepilogo del modello per verificare
model.summary()
```

```python
# get a summary of our composed model
model.summary()
```

We are now going to train our model, that is we feed the neuron with the set of training pair x, y_train from which the optimizer will find the best weights to minimize the Mean Square Error loss function (out linear regression function).

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

By looking at weights and biases we can see if the linear fit was successfull: $w_1$ represents the angular coefficient, $b$ the intercept.

```python
# return weights and biases
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

### Exercise 11.1

In order to make practice with NN, explore how does the previous linear regression depend on the number of epochs, $N_{\mathrm{epochs}}$, the number of data points $N_{\mathrm{train}}$ and on the noise $\sigma$. Try to improve the previous result operating on these parameters.

### Resolution
For this exercise is used the NN structure as explained above, using the same feature of `optimizer=sgd`,`loss=mse`,`metric=mse`

To better understand how the system depends on the parameters requested in the exercise—namely, the number of epochs, the training sample size, and the noise level—it can be helpful to observe how the loss and validation loss change as these parameters vary. The first metric, loss, is evaluated during the training process, while the validation loss reflects the error on data test that are not seen by the model during training. To explore this, we use arrays containing different values for each parameter and define a function `train_and_evaluate`, which returns both the loss and validation loss.
For the model compile

In the following analysis, several perspectives could be considered. One expected outcome is that if the validation loss increases while the training loss decreases, this may indicate overfitting, where the model becomes too tailored to the training data. Ideally, we would see both the training loss and validation loss decrease together. Additionally, I am interested in observing cases of underfitting for low values of `N_epoch`, where the model might not have had sufficient time to learn effectively. In such cases, examining the gap between the two loss curves could be particularly informative.


The values utilized to map the system are the follow in each possible combination:
$$
N_{Epoch} = ([10, 20, 30, 50, 100])
$$
$$
N_{train} =([500, 750, 1000, 1250])
$$
$$
\sigma =([0.1, 0.2, 0.5, 0.75, 1])
$$


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

def train_and_evaluate(N_train, N_epoch, sigma, m=2, b=1):
    # Genera i dati di addestramento e di validazione
    x_train = np.random.uniform(-1, 1, N_train)
    x_valid = np.random.uniform(-1, 1, 50)
    x_valid.sort()
    y_target = m * x_valid + b  # Funzione ideale lineare

    # Misure effettive con rumore gaussiano
    y_train = np.random.normal(m * x_train + b, sigma)  
    y_valid = np.random.normal(m * x_valid + b, sigma)

    # Definisce e compila il modello
    model = Sequential([
        Input(shape=(1,)),
        Dense(1)
    ])
    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

    # Addestra il modello e salva la storia
    history = model.fit(x=x_train, y=y_train, 
                        batch_size=32, epochs=N_epoch,
                        shuffle=True,
                        validation_data=(x_valid, y_valid),
                        verbose=0)
    
    return history.history['loss'], history.history['val_loss']

# Parametri per il training
N_EPOCH = np.array([10, 20, 30, 50, 100])
N_train = np.array([500, 750, 1000, 1250])
SIGMA = np.array([0.1, 0.2, 0.5, 0.75, 1])

# Esegue il training e salva i risultati per ogni combinazione di parametri
results = {}

for n_train in N_train:
    results[n_train] = {}
    for sigma in SIGMA:
        results[n_train][sigma] = {'train_loss': [], 'val_loss': []}
        for n_epoch in N_EPOCH:
            loss, val_loss = train_and_evaluate(n_train, n_epoch, sigma)
            results[n_train][sigma]['train_loss'].append(loss[-1])
            results[n_train][sigma]['val_loss'].append(val_loss[-1])
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import seaborn as sns
sns.set_theme(style='darkgrid')

# Colori per sigma
colors = ['blue', 'orange', 'green', 'red', 'purple']

# Plot per N_train fisso e N_epoch variabile
plt.figure(figsize=(35, 25))
plt.suptitle('Loss function vs N_Epoch, fixed N_train', fontsize=45)

for n_train_idx, n_train in enumerate(N_train):
    plt.subplot(2, 2, n_train_idx + 1)
    for i, sigma in enumerate(SIGMA):
        loss_values = results[n_train][sigma]['train_loss']
        val_loss_values = results[n_train][sigma]['val_loss']
        
        plt.plot(N_EPOCH, loss_values, marker='o', color=colors[i], lw=3, linestyle='-')
        plt.plot(N_EPOCH, val_loss_values, marker='x', color=colors[i], lw=3, linestyle='--')

    plt.title(f'Loss and Validation Loss vs N_EPOCH (Fixed N_train={n_train})', fontsize=35)
    plt.xlabel('N_EPOCH', fontsize=25)
    plt.ylabel('Loss', fontsize=25)
    plt.grid(True)
    plt.tick_params(axis='both', which='major', labelsize=25)

# Legenda per loss e val_loss
legend_elements = [
    Line2D([0], [0], color=colors[i], lw=3, linestyle='-', marker='o', label=f'Loss (sigma={sigma})') for i, sigma in enumerate(SIGMA)
] + [
    Line2D([0], [0], color=colors[i], lw=3, linestyle='--', marker='x', label=f'Val Loss (sigma={sigma})') for i, sigma in enumerate(SIGMA)
]

# Posiziona la legenda sotto al plot
plt.figlegend(handles=legend_elements, loc='lower center', fontsize=28, ncol=5)
plt.tight_layout(rect=[0, 0.1, 1, 0.9])
plt.subplots_adjust(bottom=0.15)
plt.show()

# Plot per N_epoch fisso e N_train variabile
plt.figure(figsize=(35, 25))
plt.suptitle('Loss function vs N_train, fixed N_Epoch', fontsize=45)

for n_epoch_idx, n_epoch in enumerate(N_EPOCH):
    plt.subplot(3, 2, n_epoch_idx + 1)
    for i, sigma in enumerate(SIGMA):
        loss_values = [results[n][sigma]['train_loss'][n_epoch_idx] for n in N_train]
        val_loss_values = [results[n][sigma]['val_loss'][n_epoch_idx] for n in N_train]
        
        plt.plot(N_train, loss_values, marker='o', color=colors[i], lw=3, linestyle='-')
        plt.plot(N_train, val_loss_values, marker='x', color=colors[i], lw=3, linestyle='--')

    plt.title(f'Loss vs N_TRAIN (Fixed N_epoch={n_epoch})', fontsize=35)
    plt.xlabel('N_TRAIN', fontsize=25)
    plt.ylabel('Loss', fontsize=25)
    plt.grid(True)
    plt.tick_params(axis='both', which='major', labelsize=25)

# Legenda per entrambi i plot
plt.figlegend(handles=legend_elements, loc='lower center', fontsize=30, ncol=5)
plt.tight_layout(rect=[0, 0.1, 1, 0.9])
plt.subplots_adjust(bottom=0.15)
plt.show()
```

# Discussion and Observations

In this plot, we can observe the comparison between `validation loss` (dashed line) and `training loss` (solid line) as different parameters change. Each color represents a different value of $\sigma$.

## Fixed Number of Training Samples
There are several aspects to note: 

1. **Loss Behavior**: As expected, for each `N_train` value, as the noise $\sigma$ increases, both the training and validation losses increase. This happens because the model finds it increasingly difficult to approximate the target function as the noise level rises.

2. **Initial Epochs**: For the first epoch values (values around 10 and 20), there is no clear trend in either the training or validation losses, likely due to underfitting. In this phase, the model hasn't yet had sufficient time to effectively learn the target function.

3. **Saturation Effect**: An interesting observation is that, particularly for low $\sigma$ values and across all epochs, the training loss reaches a plateau, showing no further improvement or degradation as the number of epochs increases.

4. **Overfitting**: Another notable point is the presence of some trend interpretated as `overfitting`: despite the training loss stabilizing, in certain cases, the validation loss begins to increase or remains constant while the training loss decreases. This indicates the model has become too adapted to the training data, rendering it less reliable on unseen data.

## Fixed Number of Epochs
In these plots, beyond the previously noted observations, we generally observe a continuous improvement in loss values, except in cases with low training samples and low epoch counts, where `valid_loss` tends to decrease or remain constant across all fixed values of `N_epoch`. Increasing the number of training samples generally improves performance, except for high training sample values, particularly when sigma is greater than 0.5, where a behavior analogous to overfitting is observed.

## Which is the Best?
It is clear that analysis with $\sigma = 0.1, 0.2$ are the best values because have the losest values of loss and valid_loss. Based on this observation, it is useless have `N_train` over $1000$ data, due to the loss values for those $\sigma$ is almost saturated.



## Simulation for one values
Below it is report an analysis of an intermediate case with $\sigma = 0.2$, $N_{train} = 1000$ and $N_{Epoch}=50$, associated also to analysis prediction. 

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
import numpy as np
import matplotlib.pyplot as plt

# Target parameters of f(x) = m*x + b
m = 2  # slope
b = 1  # intercept

# Generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 1000)
x_valid = np.random.uniform(-1, 1, 50)
x_valid.sort()
y_target = m * x_valid + b  # Ideal (target) linear function

# Noise standard deviation
sigma = 0.2 # Set sigma to 0.2 for lower noise
y_train = np.random.normal(m * x_train + b, sigma)  # Noisy training data
y_valid = np.random.normal(m * x_valid + b, sigma)  # Noisy validation data

# Define and compile the model
model = Sequential([
    Input(shape=(1,)),
    Dense(1)
])

model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
model.summary()

# Train the model
history = model.fit(x=x_train, y=y_train,
                    batch_size=32, epochs=50,
                    shuffle=True,  # Shuffle inputs at each epoch
                    validation_data=(x_valid, y_valid))

# Return weights and biases
weights, biases = model.get_weights()
print("\nWeights:", weights)
print("Biases:", biases)

# Evaluate model with noisy validation data
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)
print('\nTest loss (x_valid - y_valid):', score[0])
print('Test accuracy:', score[1])

# Evaluate model with the exact target curve (no noise)
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=0)
print('\nTest loss (x_valid - y_target):', score[0])
print('Test accuracy:', score[1])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# Plot of training e la validation loss
ax1.plot(history.history['loss'], label='Train')
ax1.plot(history.history['val_loss'], label='Validation')
ax1.set_title('Model Loss')
ax1.set_ylabel('Loss')
ax1.set_xlabel('Epoch')
ax1.legend(loc='best')

# Prediction
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)

# Plot of prediction
ax2.scatter(x_predicted, y_predicted, color='r', label='Predictions')
ax2.plot(x_valid, y_target, color='g', label='Target Function')
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()

```

Finally it is possible to notice how the convergence of the prediction data with the target one.
